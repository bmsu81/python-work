# **Visualization Of column**

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Frequency of different age groups within the customer base
column_to_visualize = 'age'
plt.figure(figsize=(10, 6))
plt.hist(df[column_to_visualize], bins=10, color='skyblue', edgecolor='black')
plt.title(f'Distribution of {column_to_visualize.capitalize()}')
plt.xlabel(column_to_visualize.capitalize())
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#  Pivot table for marital status distribution by education level
pivot_table = df.pivot_table(index='education', columns='marital', aggfunc='size', fill_value=0)
print(pivot_table)
```

```python
# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Marital Status Distribution by Education Level')
plt.xlabel('Marital Status')
plt.ylabel('Education Level')
plt.show()
```

```python
#   The distribution of account balances by loan status among the respondents
plt.figure(figsize=(8, 8))
sns.boxplot(x='housing', y='balance', data=df)
plt.title('Distribution of Account Balances by Housing Loan Status')
plt.xlabel('Housing Loan Status')
plt.ylabel('Account Balance')
plt.show()
```

```python
# How many respondents have housing loans status?
plt.figure(figsize=(8, 6))
sns.countplot(x='housing', hue='loan', data=df)
plt.title('Loan Status')
plt.xlabel('Housing Loan Status')
plt.ylabel('Count of Respondent')
plt.legend(title='Personal Loan')
plt.show()
```

```python
# Age-balance relationship based on marital status
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='balance', hue='marital', data=df, palette='Set2')
plt.title('Age vs Balance by Marital Status')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.legend(title='Marital Status', loc='upper right')
plt.show()
```

```python
# is there any variations or trends in the number of campaigns conducted over the months?
# To understand the effectivness of marketing strategy in customer engagement pattern over time.
# Grouping the data by month and counting the number of campaigns

campaigns_per_month = df.groupby('month')['campaign'].sum()

# Create a line chart
plt.figure(figsize=(10, 6))
campaigns_per_month.plot(marker='o', linestyle='-')
plt.title('Total Number of Campaigns Conducted Over Months')
plt.xlabel('Month')
plt.ylabel('Total Number of Campaigns')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Find outliers for quantitative columns using boxplots**

```python

quantitative_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Create boxplots for each quantitative column
for column in quantitative_columns:
    plt.figure(figsize=(16, 4))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

# **Create a new dataframe to find the outliers for each quantitative column**

```python
# Selecting numerical columns where  suspected outliers
numerical_columns = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']

# Creating boxplots for numerical columns
plt.figure(figsize=(16, 8))
for i, col in enumerate(numerical_columns, start=1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col], orient='v')
    plt.title(col)
plt.tight_layout()
plt.show()

# Identifying potential outliers
outliers = pd.DataFrame()
for col in numerical_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

# Creating a new DataFrame with outlier rows
outlier_rows = df[df.index.isin(outliers.index)]

# Displaying information about outlier rows
print("Information about Outlier Rows:")
print(outlier_rows)
```

# **Calculate/plot correlations between each quantitive columns**

```python
# Select quantitative columns
quantitative_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation matrix
correlation_matrix = df[quantitative_columns].corr()
print(correlation_matrix)
```

```python
# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Quantitative Columns')
plt.show()
```
